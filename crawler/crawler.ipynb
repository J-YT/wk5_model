```python
# requirements
# !pip install requests
# !pip install beautifulsoup4
# !pip install selenium
# download firefox geckodriver https://github.com/mozilla/geckodriver/releases 


import requests
from bs4 import BeautifulSoup
from selenium import webdriver


raw_html = requests.get('http://www.medisense.org.uk/do/casebank/') # get the html file from server
soup = BeautifulSoup(raw_html.text, 'html.parser') # start html parser to parse soup
# print(soup.prettify()) # visualize raw html

# to get all the names of all cases, which store in <h3>
names = soup.find_all('h3') 
for name in names:
    
    # turn name into link, i.e. http://.../.../kobe-bryant/
    split_name = name.get_text().strip().split(' ') # to split name into pieces, i.e. ['Kobe', 'Bryant']
    url_name = '-'.join(split_name) # i.e. Kobe-Bryant
    url_name = url_name.replace('’', '-') # replace all ’ with -
    url_name = url_name.replace('\'', '-') # replace all ' with -
    case_html = f'http://www.medisense.org.uk/do/casebank/{url_name}/'

    # init webdriver
    driverpath = "/Users/home/Library/Mobile Documents/com~apple~CloudDocs/Dev/siraj/wk5 midterm/scrape/geckodriver"
    driver = webdriver.Firefox(executable_path=driverpath)
    
    # get target html
    driver.get(case_html)
    
    # expand collapsed div
    driver.find_element_by_xpath("/html/body/div[1]/div/section/div[1]/div/div[1]/div/div[2]/div[1]/div/a").click()
    driver.find_element_by_xpath("/html/body/div[1]/div/section/div[1]/div/div[1]/div/div[3]/div[1]/div/a").click()

    # get text of instructions
    student_instructions = driver.find_element_by_xpath("/html/body/div[1]/div/section/div[1]/div/div[1]/div/div[1]/div[2]/div").text
    patient_instructions = driver.find_element_by_xpath("/html/body/div[1]/div/section/div[1]/div/div[1]/div/div[2]/div[2]/div").text
    examiner_instructions = driver.find_element_by_xpath("/html/body/div[1]/div/section/div[1]/div/div[1]/div/div[3]/div[2]/div").text
    
    # saved in json format
    # not done
```
